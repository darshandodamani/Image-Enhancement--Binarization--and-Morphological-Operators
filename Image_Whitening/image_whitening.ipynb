## Make the image whitening and retaining the content
    

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
# read image and convert to grayscale
image = cv2.imread('Images/example.jpg')

```

```python
#convert image to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.show()
gray.shape
```

```python
#apply gaussian blur
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
plt.imshow(blurred, cmap='gray')
plt.show()
```

```python
# make page white and remove noise from image
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(thresh, cmap='gray')
plt.show()
```

```python
# remove noise from image and make page white
kernel = np.ones((3, 3), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=2)
plt.imshow(closing, cmap='gray')
plt.show()
```

```python
# save image with white background with no noise and same title of the image
cv2.imwrite('Images/example.jpg', closing)


```
